```python
import pandas as pd
import numpy as np
```

selecting the numeric data

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasets/hepatitis/master/data/hepatitis.csv')
numeric_df = df.select_dtypes(include=np.number).copy()
numeric_df
```

Seeing how much data is missing.

```python
numeric_df.isna().sum()
```

```python
numeric_df.mean()
```

Solution:

Replacing all the missing data with the mean of the column

```python
filled_df = numeric_df.fillna(value = numeric_df.mean()).copy()
filled_df
```

Hence, Now we have no missing data.

```python
filled_df.isna().sum()
```
